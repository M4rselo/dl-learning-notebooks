### Basic Properties of Tensor Arithmetic

```python
A = torch.arange(6, dtype=torch.float32).reshape(2, 3)
B = A.clone()  # Assign a copy of A to B by allocating new memory
A, A + B, A*3, A*B, A**2, A*B == A**2
```

```python
A * B
```

```python
a = 2
X = torch.arange(24).reshape(2, 3, 4)
print("X:", X, '\n')
a + X, (a * X).shape
```
