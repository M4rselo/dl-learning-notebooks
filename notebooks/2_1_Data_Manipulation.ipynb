## Saving Memory

```python
before = id(Y)
print(before)
Y = Y + X
after = id(Y)
print(after)
after == before
```

```python
Z = torch.zeros_like(Y)
before = id(Z)
print('id(Z):', before)
Z[:] = X + Y
after = id(Z)
print('id(Z):', after)
before == after
```

```python
before = id(X)
X += Y
id(X) == before
```
